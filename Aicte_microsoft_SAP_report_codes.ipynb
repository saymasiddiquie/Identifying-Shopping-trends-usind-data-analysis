```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"C:\Users\VEDANT\OneDrive\Desktop\Datasets\shopping_trends.csv")

# 1.Distribution of customer ages
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Customer Ages', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2. Average purchase amount by product's category
avg_purchase_by_category = data.groupby('Category')['Purchase Amount (USD)'].mean().sort_values()
plt.figure(figsize=(12, 6))
avg_purchase_by_category.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Average Purchase Amount by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Gender with the highest number of purchases
purchases_by_gender = data['Gender'].value_counts()
plt.figure(figsize=(8, 5))
purchases_by_gender.plot(kind='bar', color=['skyblue', 'lightcoral'], edgecolor='black')
plt.title('Number of Purchases by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Purchases', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 4. Most commonly purchased items in each category
most_common_items = data.groupby('Category')['Item Purchased'].agg(lambda x: x.mode()[0])
print("Most Commonly Purchased Items in Each Category:")
print(most_common_items)
```

```python
# 5. Seasons or months where spending is higher
avg_purchase_by_season = data.groupby('Season')['Purchase Amount (USD)'].mean().sort_values()
plt.figure(figsize=(10, 5))
avg_purchase_by_season.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Average Purchase Amount by Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 6. Average rating by product's category
avg_rating_by_category = data.groupby('Category')['Review Rating'].mean().sort_values()
plt.figure(figsize=(10, 5))
avg_rating_by_category.plot(kind='bar', color='gold', edgecolor='black')
plt.title('Average Rating by Product Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 7. Differences between subscribed and non-subscribed customers
subscription_avg_purchase = data.groupby('Subscription Status')['Purchase Amount (USD)'].mean()
plt.figure(figsize=(8, 5))
subscription_avg_purchase.plot(kind='bar', color=['green', 'orange'], edgecolor='black')
plt.title('Average Purchase Amount by Subscription Status', fontsize=16)
plt.xlabel('Subscription Status', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 8. Most popular payment method
payment_method_counts = data['Payment Method'].value_counts()
plt.figure(figsize=(10, 6))
payment_method_counts.plot(kind='bar', color='dodgerblue', edgecolor='black')
plt.title('Most Popular Payment Methods', fontsize=16)
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 9. Do customers who use promo codes spend more than those who don't?
avg_spending_promo = data.groupby('Promo Code Used')['Purchase Amount (USD)'].mean()
plt.figure(figsize=(8, 5))
avg_spending_promo.plot(kind='bar', color=['lightblue', 'salmon'], edgecolor='black')
plt.title('Average Spending by Promo Code Usage', fontsize=16)
plt.xlabel('Promo Code Used', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 10. Correlation between product size and purchase amount
plt.figure(figsize=(8, 5))
sns.boxplot(x='Size', y='Purchase Amount (USD)', data=data, palette='viridis')
plt.title('Product Size vs. Purchase Amount', fontsize=16)
plt.xlabel('Product Size', fontsize=12)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 11. Preferred shipping type by product categories
preferred_shipping = data.groupby('Category')['Shipping Type'].value_counts(normalize=True).unstack()
preferred_shipping.plot(kind='bar', stacked=True, figsize=(12, 6), cmap='Set3')
plt.title('Preferred Shipping Type by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Proportion of Shipping Type', fontsize=12)
plt.legend(title='Shipping Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 12. Effect of discounts on purchase decisions
discount_usage = data['Discount Applied'].value_counts(normalize=True)
plt.figure(figsize=(8, 5))
discount_usage.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'], startangle=90, textprops={'fontsize': 12})
plt.title('Discount Application in Purchases', fontsize=16)
plt.ylabel('')
plt.show()
```

```python
# 13. Most popular product colors
color_popularity = data['Color'].value_counts().head(10)
plt.figure(figsize=(10, 6))
color_popularity.plot(kind='bar', color='cornflowerblue', edgecolor='black')
plt.title('Top 10 Most Popular Product Colors', fontsize=16)
plt.xlabel('Color', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 14. Average number of previous purchases
avg_prev_purchases = data['Previous Purchases'].mean()
print(f"The average number of previous purchases by customers is: {avg_prev_purchases:.2f}")

# 15. Purchase amount vs. review ratings
plt.figure(figsize=(8, 5))
sns.boxplot(x='Review Rating', y='Purchase Amount (USD)', data=data, palette='pastel')
plt.title('Purchase Amount by Review Ratings', fontsize=16)
plt.xlabel('Review Rating', fontsize=12)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#16.Average purchase amount by location
avg_purchase_by_location = data.groupby('Location')['Purchase Amount (USD)'].mean().sort_values(ascending=False)

# Plot the average purchase amount by location
plt.figure(figsize=(14, 7))
avg_purchase_by_location.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Purchase Amount by Location (Top 10)', fontsize=16)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#17. Average purchase amount by gender
avg_purchase_by_gender = data.groupby('Gender')['Purchase Amount (USD)'].mean()

# Plot the average purchase amount by gender
plt.figure(figsize=(8, 5))
avg_purchase_by_gender.plot(kind='bar', color=['lightblue', 'lightcoral'], edgecolor='black')
plt.title('Average Purchase Amount by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
